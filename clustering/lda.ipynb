```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from konlpy.tag import Okt
from tensorflow.python.keras.preprocessing.sequence import pad_sequences
from tensorflow.python.keras.preprocessing.text import Tokenizer

import warnings
warnings.filterwarnings('ignore')

import pickle
from gensim import corpora
import pyLDAvis.gensim_models
import gensim
```

```python
df = pd.read_excel(r"C:\Users\jessy\Desktop\산학연계\문장라벨링 합.xlsx")
df = df.drop([df.columns[5],df.columns[6],df.columns[7],df.columns[8]], axis=1)
df
```

```python
okt = Okt()

df['tokenized'] = 0

for i in range(0,len(df)):
    df.tokenized[i] = okt.nouns(df.sentence[i])

df
```

```python
with open('stopwords.pkl','rb') as f:
        stopwords = pickle.load(f)
        
print(stopwords)
```

```python
for i, tokenized in enumerate(df['tokenized']):
    clean_words = [] 
    for word in tokenized: 
        if word not in stopwords: #불용어 제거
            clean_words.append(word)
    df['sentence'][i] = ' '.join(clean_words) 

df['tokenized'] = 0

for i in range(0,len(df)):
    df.tokenized[i] = okt.nouns(df.sentence[i])
    
df
```

```python
dictionary = corpora.Dictionary(df.tokenized)
corpus = [dictionary.doc2bow(text) for text in df.tokenized]
df['corpus']=corpus

NUM_TOPICS = 3
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics = NUM_TOPICS, id2word=dictionary, passes=15)
topics = ldamodel.print_topics(num_words=4)

print(ldamodel.print_topics())
```

```python
pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim_models.prepare(ldamodel, corpus, dictionary)
pyLDAvis.display(vis)
```

```python
def make_topictable_per_doc(ldamodel, corpus):
    topic_table = pd.DataFrame()

    for i, topic_list in enumerate(ldamodel[corpus]):
        doc = topic_list[0] if ldamodel.per_word_topics else topic_list            
        doc = sorted(doc, key=lambda x: (x[1]), reverse=True)

        for j, (topic_num, prop_topic) in enumerate(doc):
            if j == 0:
                topic_table = topic_table.append(pd.Series([int(topic_num), round(prop_topic,4), topic_list]), ignore_index=True)
            else:
                break
    return(topic_table)

topictable = make_topictable_per_doc(ldamodel, corpus)
topictable = topictable.reset_index()
topictable.columns = ['문서 번호', '가장 비중이 높은 토픽', '가장 높은 토픽의 비중', '각 토픽의 비중']

result = pd.concat([df, topictable], axis=1)
result = result.drop(['문서 번호'], axis=1)
result
```
